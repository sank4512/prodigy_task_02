```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset= pd.read_csv("titanic_dataset.csv")
dataset.head(3)
```

```python
dataset["Age"].fillna(dataset["Age"].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.isnull().sum()
```

```python
dataset.drop(columns=['Cabin'])
```

```python
dataset.duplicated().sum()
```

```python
dataset['Sex'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
dataset['Age'].dropna().plot(kind='hist', bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
pd.crosstab(dataset['Sex'],dataset['Survived']).plot(kind='bar', stacked=True)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
dataset.boxplot(column='Age', by='Survived')
plt.title('Age Distribution by Survival')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

```python
dataset.plot.scatter(x='Age', y='Fare')
plt.title('Fare vs. Age')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
sns.pairplot(dataset[['Age', 'Fare', 'Pclass', 'Survived']].dropna())
plt.show()
```
